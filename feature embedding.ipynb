```python
df = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df["Irrigation_Need"] = df["Irrigation_Need"].map({"Low":0,"Medium":1,"High":2})
```

```python
df["Mulching_Used"] = df["Mulching_Used"].map({"No":0,"Yes":1})

df_test["Mulching_Used"] = df_test["Mulching_Used"].map({"No":0,"Yes":1})
```

```python
x = df.drop("Irrigation_Need", axis = 1 )
y = df["Irrigation_Need"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x , y , train_size= 0.90,random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

# Define categorical and numerical columns 
cat_cols = ['Soil_Type', 'Crop_Type', 'Crop_Growth_Stage', 'Season', 
            'Irrigation_Type', 'Water_Source', 'Mulching_Used', 'Region']
num_cols = ['Soil_pH', 'Soil_Moisture', 'Organic_Carbon', 'Electrical_Conductivity', 
            'Temperature_C', 'Humidity', 'Rainfall_mm', 'Sunlight_Hours', 
            'Wind_Speed_kmh', 'Field_Area_hectare', 'Previous_Irrigation_mm']

# Encode categorical features as integers
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Prepare inputs for the model
X_cat = [df[col].values for col in cat_cols]
X_num = df[num_cols].values
y = LabelEncoder().fit_transform(df['Irrigation_Need'])
```

```python
from tensorflow.keras.layers import Input, Embedding, Reshape, Concatenate, Dense, Dropout, Flatten
from tensorflow.keras.models import Model

# 1. Handle Categorical Inputs via Embeddings
input_layers = []
embedding_layers = []

for col in cat_cols:
    # Get number of unique categories
    n_unique = df[col].nunique()
    # Rule of thumb for embedding size: min(50, (n_unique + 1) // 2)
    emb_dim = min(50, (n_unique + 1) // 2)
    
    in_layer = Input(shape=(1,), name=f"input_{col}")
    emb_layer = Embedding(input_dim=n_unique, output_dim=emb_dim)(in_layer)
    emb_layer = Reshape(target_shape=(emb_dim,))(emb_layer)
    
    input_layers.append(in_layer)
    embedding_layers.append(emb_layer)

# 2. Handle Numerical Inputs
num_input = Input(shape=(len(num_cols),), name="input_numerical")
input_layers.append(num_input)

# 3. Combine Everything
combined = Concatenate()(embedding_layers + [num_input])

# 4. Add Dense Layers 
x = Dense(64, activation='relu')(combined)
x = Dropout(0.2)(x)
x = Dense(32, activation='relu')(x)
output = Dense(3, activation='softmax')(x) # Assuming 3 classes: Low, Medium, High

model = Model(inputs=input_layers, outputs=output)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
cat_col = list(x_train.select_dtypes(include=["object"]).columns)

encoder = OrdinalEncoder()

x_train[cat_col] = encoder.fit_transform(x_train[cat_col])

x_test[cat_col] = encoder.transform(x_test[cat_col])

df_test[cat_col] = encoder.transform(df_test[cat_col])
```

```python
scaler = RobustScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
df_test = scaler.transform(df_test)
```

```python
from sklearn.utils.class_weight import compute_class_weight
classes = np.unique(y_train)
weights = compute_class_weight(
    class_weight='balanced', 
    classes=classes, 
    y=y_train
)


class_weight_dict = dict(zip(classes, weights))
```

```python

model = Sequential()
model.add(Dense(128, activation="relu", input_shape=(x_train.shape[1],)))
model.add(BatchNormalization())
model.add(Dropout(0.2))


model.add(Dense(64, activation="relu", kernel_regularizer=L2(0.001)))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(32, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(20, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(3, activation="softmax"))

model.summary()
```

```python
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

lr_reduce = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6
)
```

```python
model.fit(
    x_train , y_train,
    validation_split = 0.1 , 
    epochs = 100 ,
    batch_size = 128,
    class_weight=class_weight_dict,
    callbacks=[early_stop, lr_reduce] 
)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred_df = model.predict(df_test)
```

```python
y_pred_df = np.argmax(y_pred_df, axis=1)
```

```python
reverse_map = {0: "Low", 1: "Medium", 2: "High"}
y_pred_labels = [reverse_map[int(i)] for i in y_pred_df]
```
